```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

# Chrome Browser와 Chrome Driver Version 확인
chrome_options = webdriver.ChromeOptions()

# 요청
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
url = 'https://www.koreabaseball.com/Player/Register.aspx'
driver.get(url)

pitcher_info = []


for index in range(1,11) :
    # 팀 정보 클릭
    driver.find_element(By.XPATH, f'//*[@id="cphContents_cphContents_cphContents_udpRecord"]/div[1]/ul/li[{index}]/a').click()
    time.sleep(0.7)

    # 투수들 정보가 담겨있는 tr
    pitcher_tr = driver.find_element(By.XPATH, '//*[@id="cphContents_cphContents_cphContents_udpRecord"]/div[3]/table[3]/tbody').find_elements(By.TAG_NAME, 'tr')
    time.sleep(0.7)

    # 팀 이름
    team_name = driver.find_element(By.XPATH, '//*[@id="cphContents_cphContents_cphContents_udpRecord"]/div[3]/h4').text[:-7]
    time.sleep(0.7)

    # 투수들의 상세정보들
    for pitcher in pitcher_tr :

        pitcher_dict = {}

        # 등번호
        number = pitcher.find_elements(By.TAG_NAME, 'td')[0].text
        # 이름
        name = pitcher.find_elements(By.TAG_NAME, 'td')[1].text
        # 투타 유형
        type = pitcher.find_elements(By.TAG_NAME, 'td')[2].text
        # 생년월일
        birth_date = pitcher.find_elements(By.TAG_NAME, 'td')[3].text
        # 키
        height = pitcher.find_elements(By.TAG_NAME, 'td')[4].text.split(',')[0]
        # 몸무게
        weight = pitcher.find_elements(By.TAG_NAME, 'td')[4].text.split(',')[1].strip()

        pitcher_dict = {
            '팀 이름' : team_name, 
            '등번호' : number,
            '이름' : name,
            '투타유형' : type,
            '생년월일' : birth_date,
            '키' : height,
            '몸무게' : weight
        }

        pitcher_info.append(pitcher_dict)
        time.sleep(0.7)

driver.quit()
```

```python
import pandas as pd
pitcher_df = pd.DataFrame(pitcher_info)
```
