> 필요한 데이터 : Data/rank/*.csv

```python
def teamRanking() :
    import pandas as pd
    # 사용자가 선택하는 값
    select_year = '2022'
    select_month = '04'
    team_name = '두산'

    # ------

    # 서버에서 불러와 원하는 값 추출
    team_df = pd.read_csv(f'./Data/rank/{team_name}_rank.csv')
    team_df.날짜 = pd.to_datetime(team_df.날짜)
    result_df = team_df[(team_df.날짜.dt.year == int(select_year)) & (team_df.날짜.dt.month == int(select_month))]
    result_df.loc[:, '날짜'] = result_df.날짜.astype(str)

    response = {}

    response['rank'] = []
    response['info'] = []

    for i,j in zip(result_df.to_dict()['날짜'].values(), result_df.to_dict()['순위'].values()) :
        result_dict = {
            '날짜' : i,
            '순위' : j
        }
        response['rank'].append(result_dict)


    for index in range(len(result_df.index)) :
        response['info'].append(result_df.loc[index, ['날짜', '총 경기수', '승리', '패배', '무승부', '승률', '게임차', '최근 10경기 전적', '연속 승패 현황', '홈 경기 전적', '원정 경기 전적']].to_dict())

    # return
    return response
```

```python
result = teamRanking()

print(result)
```
