```python
import pandas as pd
```

```python
# 2024 시즌 경기 일정 파일 kbo로 저장
kbo = pd.read_excel('../Data/2024kbo.xlsx')
kbo
```

```python
# 월요일은 경기가 없기 때문에 월요일 데이터는 삭제
kbo = kbo.drop(kbo[kbo['요일'] == '월'].index)
# nan 데이터는 경기가 없다 라는 데이터 임으로 - 로 변환
kbo.fillna('-',inplace=True)
kbo
```

```python
# 올스타 경기 제외
kbo = kbo.drop(kbo[(kbo['월'] == 7) & (kbo['일'].isin([5,6,7]))].index)
```

```python
# 인덱스 초기화
kbo.reset_index(drop='index',inplace=True)
```

```python
# 1차 정제된 데이터 확인
kbo
```

```python
# 구장 이름 확인
kbo.columns
```

```python
# 구장별 홈팀 원정팀으로 나누기

count = 0
kbo['서울 잠실 홈'] = 'LG'
kbo['서울 잠실 어웨이'] = '한화'
for i in kbo['서울 잠실'].str.split('-'):
    kbo.loc[count,'서울 잠실 홈'] = i[1]
    kbo.loc[count,'서울 잠실 어웨이'] = i[0]
    count += 1

count = 0
kbo['서울 고척 홈'] = 'test'
kbo['서울 고척 어웨이'] = 'test'
for i in kbo['서울 고척'].str.split('-'):
    kbo.loc[count,'서울 고척 홈'] = i[1]
    kbo.loc[count,'서울 고척 어웨이'] = i[0]
    count += 1

count = 0
kbo['인천 문학 홈'] = 'test'
kbo['인천 문학 어웨이'] = 'test'
for i in kbo['인천 문학'].str.split('-'):
    kbo.loc[count,'인천 문학 홈'] = i[1]
    kbo.loc[count,'인천 문학 어웨이'] = i[0]
    count += 1

count = 0
kbo['수원 홈'] = 'test'
kbo['수원 어웨이'] = 'test'
for i in kbo['수원'].str.split('-'):
    kbo.loc[count,'수원 홈'] = i[1]
    kbo.loc[count,'수원 어웨이'] = i[0]
    count += 1

count = 0
kbo['대전 홈'] = 'test'
kbo['대전 어웨이'] = 'test'
for i in kbo['대전'].str.split('-'):
    kbo.loc[count,'대전 홈'] = i[1]
    kbo.loc[count,'대전 어웨이'] = i[0]
    count += 1

count = 0
kbo['대구 홈'] = 'test'
kbo['대구 어웨이'] = 'test'
for i in kbo['대구'].str.split('-'):
    kbo.loc[count,'대구 홈'] = i[1]
    kbo.loc[count,'대구 어웨이'] = i[0]
    count += 1

count = 0
kbo['광주 홈'] = 'test'
kbo['광주 어웨이'] = 'test'
for i in kbo['광주'].str.split('-'):
    kbo.loc[count,'광주 홈'] = i[1]
    kbo.loc[count,'광주 어웨이'] = i[0]
    count += 1

count = 0
kbo['부산 사직 홈'] = 'test'
kbo['부산 사직 어웨이'] = 'test'
for i in kbo['부산 사직'].str.split('-'):
    kbo.loc[count,'부산 사직 홈'] = i[1]
    kbo.loc[count,'부산 사직 어웨이'] = i[0]
    count += 1

count = 0
kbo['창원 홈'] = 'test'
kbo['창원 어웨이'] = 'test'
for i in kbo['창원'].str.split('-'):
    kbo.loc[count,'창원 홈'] = i[1]
    kbo.loc[count,'창원 어웨이'] = i[0]
    count += 1

# 기존에 있던 구장 컬럼 삭제
kbo.drop(
    columns= ['서울 잠실','서울 고척','인천 문학','수원','대전','대구','광주','부산 사직','창원'],
    inplace=True
)
```
