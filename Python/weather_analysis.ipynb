# response (응답)
## - header
- resultCode : 상태 코드
- resultMsg : 결과 메시지

## - body
- dataType : XML
- items
    - item : 자세한 데이터, 딕셔너리 배열로 구성되어 있음. 
- numOfRows : 한 페이지 결과 수

```python
result_dict = {}

for index in range(len(date)) :
    # 시작 날짜
    start_date = pd.to_datetime(date[index]).strftime('%Y-%m-%d').replace('-','')
    end_date = (pd.to_datetime(date[index]) + pd.Timedelta(days=1)).strftime('%Y-%m-%d').replace('-','')

    if start_date == '2024-06-02' :
        break

    # 요청
    url = 'http://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList'
    params ={
        'serviceKey' : '3Cb8UlR8oTVFbpWz5mEhWmG3wYOPKzubI3jrcTLRzMfC49Lj4dbnDdYwe4M93C72rjxkereHtC%2B%2FeaoNq1JnXQ%3D%3D', 
        'pageNo' : '1', 
        'numOfRows' : '23', 
        'dataType' : 'XML', 
        'dataCd' : 'ASOS', 
        'dateCd' : 'HR', 
        'startDt' : start_date, 
        'startHh' : '01', 
        'endDt' : end_date, 
        'endHh' : '01', 
        'stnIds' : '108' 
    }

    response = requests.get(url, params=params)

    weather_data = xmltodict.parse(response.content)
    try :
        weather_df = pd.DataFrame(weather_data['response']['body']['items']['item'])
    except :
        continue

    # 컬럼 삭제
    weather_df.drop(columns='stnId', inplace=True)
    weather_df.drop(columns='rnum', inplace=True)
    weather_df.drop(columns='taQcflg', inplace=True)
    weather_df.drop(columns='rnQcflg', inplace=True)
    weather_df.drop(columns='wsQcflg', inplace=True)
    weather_df.drop(columns='wdQcflg', inplace=True)
    weather_df.drop(columns='hmQcflg', inplace=True)
    weather_df.drop(columns='pv', inplace=True)
    weather_df.drop(columns='td', inplace=True)
    weather_df.drop(columns='paQcflg', inplace=True)
    weather_df.drop(columns='ps', inplace=True)
    weather_df.drop(columns='psQcflg', inplace=True)
    weather_df.drop(columns='ssQcflg', inplace=True)
    weather_df.drop(columns='clfmAbbrCd', inplace=True)
    weather_df.drop(columns='lcsCh', inplace=True)
    weather_df.drop(columns='ts', inplace=True)
    weather_df.drop(columns='tsQcflg', inplace=True)
    weather_df.drop(columns='m005Te', inplace=True)
    weather_df.drop(columns='m01Te', inplace=True)
    weather_df.drop(columns='m02Te', inplace=True)
    weather_df.drop(columns='m03Te', inplace=True)
    weather_df.drop(columns='dsnw', inplace=True)
    weather_df.drop(columns='hr3Fhsc', inplace=True)
    weather_df.drop(columns='dc10Tca', inplace=True)
    weather_df.drop(columns='dc10LmcsCa', inplace=True)
    weather_df.drop(columns='gndSttCd', inplace=True)
    weather_df.drop(columns='wd', inplace=True)

    # 보통 경기가 시작되는 시간대 사이 추출
    weather_df = weather_df[11:19]
    weather_df.columns = ['시간', '위치', '기온', '강수량', '풍속', '습도', '기압', '일조', '일사', '시정', '현상번호']

    # str type
    weather_df['기온'].fillna('0', inplace=True)
    weather_df['기온'] = weather_df['기온'].astype(float)

    weather_df['강수량'].fillna('0', inplace=True)
    weather_df['강수량'] = weather_df['강수량'].astype(float)

    weather_df['풍속'].fillna('0', inplace=True)
    weather_df['풍속'] = weather_df['풍속'].astype(float)

    weather_df['습도'].fillna('0', inplace=True)
    weather_df['습도'] = weather_df['습도'].astype(float)

    weather_df['기압'].fillna('0', inplace=True)
    weather_df['기압'] = weather_df['기압'].astype(float)

    weather_df['일조'].fillna('0', inplace=True)
    weather_df['일조'] = weather_df['일조'].astype(float)

    weather_df['일사'].fillna('0', inplace=True)
    weather_df['일사'] = weather_df['일사'].astype(float)

    weather_df['시정'].fillna('0', inplace=True)
    weather_df['시정'] = weather_df['시정'].astype(float)

    weather_df['현상번호'].fillna('0', inplace=True)
    weather_df['현상번호'] = weather_df['현상번호'].astype(int)

    weather_df.reset_index(drop='index', inplace=True)

    weather_dict = weather_df.to_dict()

    # if weather_df.to_dict()['시간'][0][:10] not in result_dict.keys() :
    result_dict[weather_df.to_dict()['시간'][0][:10]] = weather_dict
```
