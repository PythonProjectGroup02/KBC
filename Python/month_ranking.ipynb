```python
def teamRanking() :
    import pandas as pd
    # 사용자가 선택하는 값
    select_year = '2022'
    select_month = '04'
    team_name = '두산'

    # ------

    # 서버에서 불러와 원하는 값 추출
    team_df = pd.read_csv(f'./Data/rank/{team_name}_rank.csv')
    team_df.날짜 = pd.to_datetime(team_df.날짜)
    result_df = team_df[(team_df.날짜.dt.year == int(select_year)) & (team_df.날짜.dt.month == int(select_month))]
    result_df.loc[:, '날짜'] = result_df.날짜.astype(str)

    response = {}

    response['rank'] = []
    response['info'] = []

    for i,j in zip(result_df.to_dict()['날짜'].values(), result_df.to_dict()['순위'].values()) :
        result_dict = {
            '날짜' : i,
            '순위' : j
        }
        response['rank'].append(result_dict)


    for index in range(len(result_df.index)) :
        response['info'].append(result_df.loc[index, ['날짜', '총 경기수', '승리', '패배', '무승부', '승률', '게임차', '최근 10경기 전적', '연속 승패 현황', '홈 경기 전적', '원정 경기 전적']].to_dict())

    # return
    return response
```

```python
result = teamRanking()

print(result)
```

```python
import pandas as pd

team_name_arr = ["두산", "롯데", "삼성", "키움", "한화", "KIA", "KT", "LG", "NC", "SSG"]

select_month = 4
select_year = 2024

response = {}

for teamName in team_name_arr :
    team_df = pd.read_csv(f'./Data/rank/{teamName}_rank.csv')
    team_df = team_df.iloc[:, :3]
    team_df.columns = ['date', 'team', 'rank']
    team_df.date = pd.to_datetime(team_df.date)
    result_df = team_df[(team_df.date.dt.year == int(select_year)) & (team_df.date.dt.month == int(select_month))]
    result_df.date = result_df.date.astype(str)

    if teamName not in response.keys() :
        response[teamName] = []

    for i in range(len(result_df)) :
        response[teamName].append(result_df.iloc[i,:].to_dict())

response
```
