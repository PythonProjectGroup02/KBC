```python
import pandas as pd
```

```python
# 2024 시즌 경기 일정 파일 kbo로 저장
kbo = pd.read_excel('Data/2024kbo.xlsx')
kbo
```

```python
# 월요일은 경기가 없기 때문에 월요일 데이터는 삭제
kbo = kbo.drop(kbo[kbo['요일'] == '월'].index)
# nan 데이터는 경기가 없다 라는 데이터 임으로 - 로 변환
kbo.fillna('-',inplace=True)
kbo
```

```python
# 올스타 경기 제외
kbo = kbo.drop(kbo[(kbo['월'] == 7) & (kbo['일'].isin([5,6,7]))].index)
```

```python
# 인덱스 초기화
kbo.reset_index(drop='index',inplace=True)
```
