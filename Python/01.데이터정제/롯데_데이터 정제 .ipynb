```python
# 롯데 전적 관련 데이터 가져오기
lotte_rank = pd.read_csv('../Data/rank/롯데_rank.csv')
lotte_rank
```

```python
# kiwoom 기준으로 내부 조인 병합
lotte_new = lotte_rank.merge(lotte, on="날짜", how="inner")
lotte_new
```
