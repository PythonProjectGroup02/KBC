```python
# 라이브러리
import pandas as pd
```

```python
# 차트 경고 메세지 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 롯데 관중 수 없는 데이터 가져오기
kt_no_crowd = pd.read_csv('../Data/kt.csv')
kt_no_crowd
```

```python
kt_no_crowd['구장'].unique()
```

```python
# 구장이 사직인 데이터만 가져오기
kt_no_crowd = kt_no_crowd[kt_no_crowd['구장'] == '구장 : 수원']
kt_no_crowd
```

```python
# 날짜 데이터 요일 제거
kt_no_crowd['날짜'] = kt_no_crowd['날짜'].str[:10].astype('datetime64[ns]')
kt_no_crowd
```

```python
# 정보 확인하기
kt_no_crowd.info()
```

```python
# 2024년 06월 이전 데이터만 가져오기
kt_no_crowd = kt_no_crowd[kt_no_crowd['날짜'] < '2024-06-01']
kt_no_crowd
```

```python
# 롯데 관중 수 있는 데이터 가져오기
kt_crowd = pd.read_csv('../Data/crowd/kt.csv')
kt_crowd
```

```python
# 정보확인
kt_crowd.info()
```

```python
# 날짜 데이터 타입 변환
kt_crowd['날짜'] = kt_crowd['날짜'].astype('datetime64[ns]')
kt_crowd
```

```python
# 정보 확인
kt_crowd.info()
```

```python
# 2024년 06월 이전 데이터만 가져오기
kt_crowd = kt_crowd[kt_crowd['날짜'] < '2024-06-01']
kt_crowd
```

```python
# 날짜가 중복된 데이터 확인
kt_crowd[kt_crowd.날짜.duplicated()]
```

> 2023-09-09 2개 경기가 있는거 확인.(실제 데이터 확인 결과)

```python
kt_crowd[kt_crowd.날짜 == '2023-10-04']
```

```python
# 날짜가 중복된 데이터 확인
kt_no_crowd[kt_no_crowd.날짜.duplicated()]
```

```python
# kiwoom_crowd 기준 병합
kt = \
    pd.merge(
        kt_crowd, kt_no_crowd,
        on='날짜',
        how='inner'
        )
kt
```

```python
# 날짜 중복된 데이터 확인
kt[kt.날짜 == '2023-10-04']
```

```python
# lotte_crowd와 lotte_no_crowd은 날짜 기준, 시간 오름차순 정렬
# 140, 141행은 거짓
kt = kt.drop(index=[139,140])
kt = kt.reset_index(drop=True)
kt
```

```python
# 롯데 전적 관련 데이터 가져오기
kt_rank = pd.read_csv('../Data/rank/KT_rank.csv')
kt_rank
```

```python
kt_rank.info()
```

```python
# 날짜기준으로 병합하기 위해 타입 변환
kt_rank['날짜'] = kt_rank['날짜'].astype('datetime64[ns]')
kt_rank
```

```python
kt_rank.info()
```

```python
# kiwoom 기준으로 내부 조인 병합
kt_new = kt_rank.merge(kt, on="날짜", how="inner")
kt_new
```

```python
# 데이터 저장
kt_new.to_csv('../Data/final_data/kt_new.csv', index=None)
```
